# K-smooth numbers

*Based on the exercises by Dr Mike X Cohen*

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
maxN = 10000
largestPF = np.zeros(maxN+1, dtype=int)*np.nan #replace zeros with nan

for i in range(2, maxN+1):
  largestPF[i] = np.max(sym.primefactors(i))

plt.plot(largestPF, 'k.', markersize=0.3)
plt.xlabel('Number')
plt.ylabel('Largest prime factor')
plt.show()
```

Now, let's count the number of occurences of max prime factors

```python
uniqueMP = np.unique(largestPF)

counts = np.zeros(len(uniqueMP), dtype=int)

for i,u in enumerate(uniqueMP):
  counts[i] = np.sum(largestPF==u)
```

```python
plt.plot(uniqueMP, counts,'k.', markersize=3)
plt.title('%s is the modal max prime factor' %uniqueMP[np.argmax(counts)])
plt.show()
```

```python
# 5-smooth numbers in the given range
# 5-smooth numbers are also known as regular numbers
numberlist = np.arange(maxN+1)
print('\x1b[1;37;40m' + 'There are %s numbers that are 5-smooth.' %len(numberlist[largestPF<=5]) + '\x1b[0m')
print(numberlist[largestPF<=5])
```

```python
countSmooth = np.zeros(len(uniqueMP))
for i,u in enumerate(uniqueMP):
  countSmooth[i] =  len(numberlist[largestPF<=u])

plt.plot(uniqueMP, countSmooth, 'k.', markersize=3)
plt.xlabel('Smooth number')
plt.ylabel('Number of numbers')
plt.title('K-smooth number frequencies')
plt.show()
```
